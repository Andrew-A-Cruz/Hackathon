```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
import random
```

```python
product_types =[
        'Commercial Auto & Trucking',
        'Cyber',
        'General Liability',
        'Management & Professional Liability',
        'Property',
        'Business Owner Policy',
        'Surety Bonds',
        'Workers Compensation',
        'Car',
        'Home',
        'Renters',
        'Condo',
        'Landlord',
        'Boat & Yacht',
        'Flood',
        'Motorcycle',
        'Travel',
        'Pet',
        'Weddings & Events',
        'Umbrella',
        'Other'
    ]
data = pd.DataFrame()
for i in range(1,1001):
    product_id = i
    product_type = product_types[random.randint(0, 20)]
    popularity_scale = random.randint(1, 10)
    durability_scale = random.randint(1, 10)
    price_scale = random.randint(1, 10)
    tempData = pd.DataFrame(
        [[
            product_id,
            product_type,
            popularity_scale,
            durability_scale,
            price_scale
        ]],
        columns = [
            'product_id',
            'product_type',
            'popularity_scale',
            'durability_scale',
            'price_scale'
        ]
    )
    data = pd.concat([tempData, data])
data.head()
```

```python
nn_data = data.drop(['popularity_scale'], axis=1)
dummy_df = pd.get_dummies(nn_data['product_type'])
nn_data = pd.concat([nn_data, dummy_df], axis=1)
nn_data = nn_data.drop('product_type', axis=1)
model = NearestNeighbors(n_neighbors=5)
nn_data.head()
```

```python
model.fit(nn_data)
```

```python
# product to reccommend on randomized for example
row_to_predict = nn_data.iloc[random.randint(0, 1000)]
prediction_indices = model.kneighbors([row_to_predict])[1][0]
nearest_neighbors = nn_data.iloc[prediction_indices]
```
