```python
import pandas as pd
import matplotlib as mp
import random
import datetime
import json
import home
```

```python
homeworlds = open('home_worlds.json')
homeworlds_data = json.load(homeworlds)

name_arr = []
for i in homeworlds_data:
    name_arr.append(i['name'])
len(name_arr)
```

```python
unit_types =[
    "stormtrooper", 
    "tie_fighter", 
    "at-st", 
    "x-wing", 
    "resistance_soldier", 
    "at-at", 
    "tie_silencer", 
    "unknown"
]

empire_or_resistance = ['empire', 'resistance']
empire_or_resistance[random.randint(0, 1)]
```

```python
data = pd.DataFrame()
for i in range(1,1001):
    timestamp = datetime.datetime.now()
    unit_id = i
    unit_type = unit_types[random.randint(0, 7)]
    alliance = empire_or_resistance[random.randint(0, 1)]
    location_x = random.randint(0, 10)
    location_y = random.randint(0, 10)
    destination_x = random.randint(0, 10)
    destination_y = random.randint(0, 10)
    homeworld = name_arr[random.randint(0, 41)]
    tempData = pd.DataFrame(
        [[
            timestamp, 
            unit_id,
            unit_type,
            alliance,
            location_x,
            location_y,
            destination_x,
            destination_y,
            homeworld
        ]],
        columns = [
            'timestamp', 
            'unit_id',
            'unit_type',
            'empire_or_resistance', 
            'location_x',
            'location_y',
            'destination_x',
            'destination_y',
            'homeworld'
        ]
    )
    data = pd.concat([tempData, data])
data
```

```python
data.to_csv('stdata.csv', index=False)
```
